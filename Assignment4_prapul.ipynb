```python
from matplotlib import pyplot as plt
import numpy as np
x = np.linspace(1,50,100)       #setting values 1 to 50
y = 2+(3*x)      #line function
plt.plot(x, y, 'b-', label='y=2+(3*x)')   #plotting line graph
plt.title('Graph of y=2x+1')     #title of the graph
plt.xlabel('x')   #name for x label
plt.ylabel('y')   #name for y label
plt.legend(loc='upper left')  #displaying line function top left corner
plt.grid()    #plot grid lines
plt.show()
```
